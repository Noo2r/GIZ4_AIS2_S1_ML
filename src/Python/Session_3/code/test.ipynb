```python
"""
calculate Factorial of number using recursion
"""
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
factorial(5)
```

```python
def prime(num):
    for i in range(2, num):
        if num % i == 0:
            return f"{num} is not a prime number"
    return f"{num} is a prime number"
prime(4)    
```

```python
"""
this function will return all common divisors between two numbers
a: int
b: int
return list of common divisors 
type of ruturn is list  
"""
def common_divisor(a, b):
    divisors = []
    for i in range(1, min(a, b) + 1):
        if a % i == 0 and b % i == 0:
            divisors.append(i)
    return divisors
```

```python
common_divisor(12, 18)
```
