```python
def reverse_vocab(s):
    """Reverses the vocabulary of the given string.

    Args: string
    returns: string but reversed word by word
    """
    words = s.split()
    reversed_words = words[::-1]
    return ' '.join(reversed_words)
```

```python
reverse_vocab("Hello world from Python")  
```

```python
import random 
def generate_password(lenght):
    """
    Generates a random password of given length.
    Args: lenght (int)
    returns: string
    """

    characters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789!@#$%^&*()"
    password = ''.join(random.choice(characters) for _ in range(lenght))
    return password
```

```python
generate_password(12)
```

```python
def count_unique_words(s):
    """
    Counts the number of unique words in a list.
    Args: list of words (string)
    returns: list for each unique word its count (int)
    """
    unique_word = set(s)
    word_count = {word: s.count(word) for word in unique_word}
    return word_count
```

```python
words = ["Welcome", "Ali", "Hi", "Ali", "No", "Hi", "No", "Ali", "No", "Ali"]
count_unique_words(words)
```

```python
import random

responses = {
    "hello": ["Hi there!", "Hello!", "Greetings!"],
    "how are you?": ["I'm good, thank you!", "Doing well"],
    "what is your name?": ["I'm a chatbot.", "I don't have a name."],
    "what are you doing?": ["Just chatting with you", "Helping you out!"],
    "goodbye": ["See you later!", "Goodbye!", "Take care!"],
    "default": ["i am sorry, I don't understand.", "Can you rephrase that?"]
}

def get_response(user_input):
    """
    Generates a response based on user input.
    Args: user_input (string)
    returns: response (string)
    """
    if user_input in responses:
        response = random.choice(responses[user_input])
    else:
        response = random.choice(responses["default"])
    print("Chatbot:", response)

def chatbot():
    """
    Generates a chatbot response based on user input.
    Args: user_input (string)
    returns: response (string)
    """
    print("Chatbot: hello! how can I help you today? (Type 'exit' to quit)")
    while True:
        user_input = input("You: ").lower()
        if user_input == 'exit':
            print("Chatbot: Goodbye!")
            break
        else:
            get_response(user_input)
```

```python
chatbot()
```
